```python
import cv2
```

```python
# Open the video file
cap = cv2.VideoCapture("./videos/S10_cars_video.mp4")
cap
```

```python
cap.read() # returns a tuple (ret, frame)
```

```python
while True: # reads the video frame by frame
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow("Frame", frame)
    if cv2.waitKey(10) == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
cap.get(cv2.CAP_PROP_FRAME_COUNT) # Total number of frames in the video
```

```python
cap.get(cv2.CAP_PROP_FPS) # The video is 30 frames per second
```

```python
img = cv2.imread("./images/A-Cat.jpg")
```

```python
brighter = img + 50
cv2.imshow("Brighter Image", brighter)
cv2.waitKey(10)
cv2.destroyAllWindows()
```

```python
brighter2 = img + 50.0
cv2.imshow("Brighter Image", brighter2)
cv2.waitKey(10)
cv2.destroyAllWindows()
```

```python
brighter3 = cv2.add(img, 50) # This is a better way to brighten the image
cv2.imwrite('./images/A-Cat-brighter.jpg', brighter3) 
```

```python
darker = cv2.subtract(img, 50)
cv2.imwrite('./images/A-Cat-darker.jpg', darker) 
```

```python
multiplied = cv2.multiply(img, 1.5) # for contrast enhancement
cv2.imwrite('./images/A-Cat-multiplied.jpg', multiplied) 
```

```python
divided = cv2.divide(img, 1.5) # for contrast reduction
cv2.imwrite('./images/A-Cat-divided.jpg', divided) 
```

```python
_, frame = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY) 
cv2.imwrite('./images/A-Cat-thresholded.jpg', frame)
```

```python
lenna = cv2.imread("./images/Lenna.png")
lucy = cv2.resize(img, lenna.shape[:2])
```

```python
blended = cv2.addWeighted(lenna, 0.8, lucy, 0.2, 1)
cv2.imwrite('./images/Lenna-Lucy-blended.jpg', blended)
```
