```python
cv2.imshow('frame', image) # opens an image in a window named 'frame'
cv2.waitKey(0) # waits for a key press to close the window, 0 means wait indefinitely (it does not work in Jupyter notebooks, use in a .py file)
cv2.destroyAllWindows() # closes all windows (it does not work in Jupyter notebooks, use in a .py file)
```

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # converts the image to grayscale
cv2.imwrite('./images/A-Cat-Grey.jpg', grey_image) 
```

```python
# cv2.resize(image, (width, height))
resized_image = cv2.resize(image, (900, 500)) # resizes the image to 900x500 pixels
cv2.imwrite('./images/A-Cat-Resized.jpg', resized_image)
```

```python
# image[height_start:height_end, width_start:width_end]
cropped_image = image[700:1500, 700:2500] # crops the image to the region of interest (ROI) defined by the pixel coordinates (y1:y2, x1:x2)
cv2.imwrite('./images/A-Cat-Cropped.jpg', cropped_image)
```

```python
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE) # rotates the image 90 degrees clockwise
cv2.imwrite('./images/A-Cat-Rotated.jpg', rotated_image)
```

```python
# custom rotation
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
point = (w // 4 + 100, h // 3 + 100)

# matrix = cv2.getRotationMatrix2D(center, 45, 1)
matrix = cv2.getRotationMatrix2D(point, 55, 0.5)
custom_rotated_image = cv2.warpAffine(image, matrix, (w, h))
cv2.imwrite('./images/A-Cat-Custom-Rotated.jpg', custom_rotated_image)
```

```python
#flip
flipped_image = cv2.flip(image, -1) # flips the image horizontally (1 for horizontal, 0 for vertical, -1 for both)
cv2.imwrite('./images/A-Cat-Flipped.jpg', flipped_image)
```
