#### Broadcasting

Given a=np.arange(3).reshape(3, 1) and b=np.arange(3):

- Predict the result of a + b without running the code.
    + It's going to be a 3x3 matrix as NumPy will apply broadcasting:
      [
        [0, 1, 2],
        [1, 2, 3],
        [2, 3, 4],
      ]

- Explain the broadcasting rules at play.
    + It compares the shape of a and b from right to left and here a's shape is 2D (3,1) but b's shape is 1D (3,).
    + Numpy will now expand b to treat as (1, 3). 
    + Two dimentions are only compatible:
        - They are same/equal OR one of them is 1.


- Create the arrays and run the results
    + array([[0, 1, 2],
       [1, 2, 3],
       [2, 3, 4]])

- Implement a solution that would not rely on broadcasting (using np.tile or similar).
    ```python
    a_without_broadcating = np.tile(a, (1,3))
    b_without_broadcating = np.tile(b, (3,1))

    a_without_broadcating+b_without_broadcating
    ```

````python
# With broadcasting
a = np.arange(3).reshape(3,1) 
b = np.arange(3)

a+b
````

````python
# Without broadcasting (with np.tile())
a_without_broadcating = np.tile(a, (1,3))
b_without_broadcating = np.tile(b, (3,1)) # np.tile() repeats an array multiple times to create a larger array/matrix of perticular shape.

a_without_broadcating+b_without_broadcating
````
